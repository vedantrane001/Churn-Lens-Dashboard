```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/raw/WA_Fn-UseC_-Telco-Customer-Churn.csv')

# Convert TotalCharges to numeric
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Handle missing values (safe way)
df['TotalCharges'] = df['TotalCharges'].fillna(0)

# Drop unnecessary column
df.drop('customerID', axis=1, inplace=True)
```

```python
df['Churn'].value_counts()
```

```python
sns.countplot(x='Churn', data=df)
plt.title('Churn Distribution')
plt.show()
```

```python
sns.countplot(x='gender', data=df)
plt.title('Gender Distribution')
plt.show()
```

```python
sns.countplot(x='Contract', data=df)
plt.title('Contract Type Distribution')
plt.xticks(rotation=15)
plt.show()
```

```python
sns.histplot(df['tenure'], bins=30, kde=True)
plt.title('Tenure Distribution')
plt.show()
```

```python
sns.histplot(df['MonthlyCharges'], bins=30, kde=True)
plt.title('Monthly Charges Distribution')
plt.show()
```

```python
sns.histplot(df['TotalCharges'], bins=30, kde=True)
plt.title('Total Charges Distribution')
plt.show()
```

```python
sns.countplot(x='Contract', hue='Churn', data=df)
plt.title('Churn vs Contract Type')
plt.xticks(rotation=15)
plt.show()
```

```python
sns.countplot(x='InternetService', hue='Churn', data=df)
plt.title('Churn vs Internet Service')
plt.show()
```

```python
sns.boxplot(x='Churn', y='MonthlyCharges', data=df)
plt.title('Monthly Charges vs Churn')
plt.show()
```

```python
sns.boxplot(x='Churn', y='tenure', data=df)
plt.title('Tenure vs Churn')
plt.show()
```

```python
df.to_csv('../data/cleaned_telco_churn.csv', index=False)
```

```python
Exploratory Data Analysis Summary
Churn rate is higher among month-to-month contract customers
Customers with higher monthly charges are more likely to churn
Short-tenure customers show higher churn probability
Internet service type influences churn behavior
These insights indicate that pricing strategy, contract duration, and service type play a crucial role in customer retention.
```
